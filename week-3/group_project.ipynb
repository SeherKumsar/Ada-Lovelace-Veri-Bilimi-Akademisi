```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import colors as mcolors
from matplotlib import patches as mpatches
```

```python
base_url = "../data/"
data = base_url + "birds.csv"
```

```python
birds = pd.read_csv(data)
```

```python
birds.head()
```

```python
birds.info()
```

```python
family_counts = birds["Family"].value_counts()
print(family_counts)
print(type(family_counts))
```

```python
family_counts = birds["Family"].value_counts()
# Seri -> Dataframe
family_counts_df = pd.DataFrame({'Family': family_counts.index, 'SpeciesCount': family_counts.values})
family_counts_df
```

```python
family_counts = birds["Family"].value_counts()
# Seri -> Dataframe
family_counts_df = pd.DataFrame({'Family': family_counts.index, 'SpeciesCount': family_counts.values})

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Family', y='SpeciesCount', data=family_counts_df)
plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('Number of Species')
plt.title('Number of Species per Family')
plt.show()
```

```python
plt.barh(y = birds["Family"].value_counts().sort_values(ascending=False).index, 
         width =birds["Family"].value_counts().sort_values(ascending=False).values,
         figure = plt.figure(figsize=(10, 10)))
```

```python
genus_max_wingspan_df = birds.groupby('Genus')['MaxWingspan'].max().reset_index()
genus_max_wingspan_df
```

```python
#####cinsine göe max kanat uzunlukları ########
genus_max_wingspan_df = birds.groupby('Genus')['MaxWingspan'].max().reset_index()

# Plotting
plt.figure(figsize=(120, 90))
sns.barplot(x='Genus', y='MaxWingspan', data=genus_max_wingspan_df)
plt.xticks(rotation=90)
plt.xlabel('Kuş Cinsi')
plt.ylabel('Maksimum Kanat Uzunluğu')
plt.title('Kuş Cinsine Göre Maksimum Kanat Uzunluğu')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x = birds['Family'],
            y = birds['MinBodyMass']
            )

plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('MinBodyMass')
plt.title('MinBodyMass per Family')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x = birds['Family'],
            y = birds['ConservationStatus']
            )

plt.xticks(rotation=90)
plt.xlabel('Family')
plt.ylabel('')
plt.title('ConservationStatus per Family')
plt.show()
```

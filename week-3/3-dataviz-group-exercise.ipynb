```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (12,6)
```

```python
birds_url = "../data/birds.csv"
birds = pd.read_csv(birds_url)
```

```python
birds.head()
```

```python
birds.info()
```

```python
# Visualize birds wingspan
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
wingspan = birds.MaxWingspan 
wingspan.plot()
plt.show()
```

```python
# Check outliers
birds[birds["MaxWingspan"] > 500]
```

```python
# Label outliers
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both', which='both', labelbottom=False, bottom=False)

for i in range(len(birds)): # Her bir veri noktas iin
    x = birds['Name'][i] # x eksenindeki konumunu belirle
    y = birds['MaxWingspan'][i] # y eksenindeki konumunu belirle
    plt.plot(x, y, "o", color="steelblue") # Veri noktasn iz
    # Outliers deerleri belirginletir ve etiketle
    if birds['MaxWingspan'][i] > 500: # Eer deer 500'den bykse
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=10,) # Deerin zerine adn etiketle
    
plt.show()
```

500 santimetre zerindeki deerlerin yanna ku isimlerini eklemek iin bir fonksiyon yazdk.

Outlier deerleri belirginletirdikten sonra, bu noktalar veri setimizden karalm. Veri setimizdeki aykr deerleri kardktan sonra, tekrar grselletirelim.

```python
# 1- Remove outliers in plot

plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)): # Her bir veri noktas iin
    x = birds['Name'][i] # x eksenindeki konumunu belirle
    y = birds['MaxWingspan'][i] # y eksenindeki konumunu belirle
    if birds['Name'][i] not in ['Bald eagle', 'Prairie falcon']: # Eer deer 500'den byk olan deerlerden biri deilse
        plt.plot(x, y, "o", color="steelblue") # Veri noktalarn iz
plt.show()
```

```python
# 2- Remove outliers before plot
birds = birds[birds["MaxWingspan"] < 500] # 500'den kk deerleri al
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
wingspan = birds.MaxWingspan # wingspan deikenine MaxWingspan stununu atayarak 500'den kk deerleri izdir
wingspan.plot(kind="bar")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (12,6)
```

```python
birds_url = "../data/birds.csv"
birds = pd.read_csv(birds_url)
```

```python
birds.head()
```

```python
birds.info()
```

```python
# Visualize birds wingspan
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
wingspan = birds.MaxWingspan 
wingspan.plot()
plt.show()
```

```python
# Check outliers
birds[birds["MaxWingspan"] > 500]
```

```python
# Label outliers
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both', which='both', labelbottom=False, bottom=False)

for i in range(len(birds)): # Her bir veri noktası için
    x = birds['Name'][i] # x eksenindeki konumunu belirle
    y = birds['MaxWingspan'][i] # y eksenindeki konumunu belirle
    plt.plot(x, y, "o", color="steelblue") # Veri noktasını çiz
    # Outliers değerleri belirginleştir ve etiketle
    if birds['MaxWingspan'][i] > 500: # Eğer değer 500'den büyükse
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=10,) # Değerin üzerine adını etiketle
    
plt.show()
```

500 santimetre üzerindeki değerlerin yanına kuş isimlerini eklemek için bir fonksiyon yazdık.

Outlier değerleri belirginleştirdikten sonra, bu noktaları veri setimizden çıkaralım. Veri setimizdeki aykırı değerleri çıkardıktan sonra, tekrar görselleştirelim.

```python
# 1- Remove outliers in plot

plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)): # Her bir veri noktası için
    x = birds['Name'][i] # x eksenindeki konumunu belirle
    y = birds['MaxWingspan'][i] # y eksenindeki konumunu belirle
    if birds['Name'][i] not in ['Bald eagle', 'Prairie falcon']: # Eğer değer 500'den büyük olan değerlerden biri değilse
        plt.plot(x, y, "o", color="steelblue") # Veri noktalarını çiz
plt.show()
```

```python
# 2- Remove outliers before plot
birds = birds[birds["MaxWingspan"] < 500] # 500'den küçük değerleri al
```

```python
birds.info()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
wingspan = birds.MaxWingspan # wingspan değişkenine MaxWingspan sütununu atayarak 500'den küçük değerleri çizdir
wingspan.plot(kind="bar")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import colors as mcolors
import seaborn as sns
from matplotlib import patches as mpatches
```

```python
df = pd.DataFrame(
    {
        "Group": ["A", "B", "C", "D", "E"],
        "value": [1, 2, 3, 4, 5]
    }
)
```

```python
plt.barh(y=df.Group, width=df.value) # color=df.color
```

```python
df = df.sort_values(by='value')
plt.barh(y=df.Group, width=df.value)
```

```python
df = df.sort_values(by='value')
plt.barh(y=df.Group, 
         width=df.value, 
         color=['red', 'blue', 'orange'], 
         edgecolor='black',
         linewidth=3
        )
```

```python
df = df.sort_values(by='value')
my_plot = plt.barh(
        y=df.Group, 
        width=df.value, 
        color=mcolors.TABLEAU_COLORS,
        edgecolor='black',
        linewidth=3
        )
plt.bar_label(container=my_plot, padding=10)
plt.show()
```

my_plot değişkenine atama yapılarak my_plot.bar_label()

```python
tips = sns.load_dataset("tips") # load_dataset ile bir çok veri setine erişebilirsiniz
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.barplot(
        x="day", 
        y="total_bill", 
        data=tips
        )
```

```python
sns.set(style="darkgrid")
sns.barplot(
        x="total_bill", 
        y="day", 
        data=tips,
        color="cyan"
        )
plt.figure(figsize=(18, 8)) # figsize ile grafik boyutunu değiştirebiliriz
```

```python
# My example
sns.barplot(
        x="total_bill",
        y="smoker",
        hue="sex",
        data=tips,
        palette="dark:pink" # color:pink is deprected
        )
plt.figure(figsize=(16, 8))
```

```python
# Aslıhan's example
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
total = tips.groupby('day')['total_bill'].sum().reset_index()
```

```python
total
```

```python
smoker = tips[tips['smoker'] == 'Yes']
smoker
```

```python
bar2 = sns.barplot(
    x='day', 
    y='total_bill', 
    data=smoker, 
    estimator=sum,
    errorbar=None,
    color='lightblue'
    )
```

```python
bar1 = sns.barplot(x='day', y='total_bill', data=total, color='darkblue')
bar2 = sns.barplot(
    x='day', 
    y='total_bill', 
    data=smoker, 
    estimator=sum,
    errorbar=None,
    color='lightblue'
    )
top_bar = mpatches.Patch(color='darkblue', label='non-smoker')
bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes')
plt.legend(handles=[top_bar, bottom_bar])
plt.show()
```

```python
df_iris = sns.load_dataset('iris')
df_iris.head()
```

```python
sns.histplot(data=df_iris, x='sepal_length', kde=True) # kde=True ile yoğunluk grafiğini çizgi ile gösterir
```

```python
# sns.displot(data=df_iris, x='sepal_length', kde=True)
```

```python
sns.displot(df_iris["sepal_length"])
# displot is deprecated, please use displot instead
```

```python
sns.regplot(x=df_iris["sepal_length"],
            y=df_iris["sepal_width"],
            line_kws={
                "color":"red",
                "alpha":0.6,
                "lw":5
            }
            )
```

```python
sns.regplot(x=df_iris["sepal_length"],
            y=df_iris["sepal_width"],
            marker="+", # marker ile nokta şeklini string olarak değiştirebiliriz
            line_kws={
                "color":"red",
                "alpha":0.6,
                "lw":5
            },
            scatter_kws={
                "color":"darkblue",
                "s":50 # s ile nokta büyüklüğünü değiştirebiliriz
            }
            )
```

```python
sns.lmplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=df_iris)
```

```python
sns.lmplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=df_iris,
    fit_reg=False) # fit_reg=False ile regresyon çizgisini kaldırabiliriz)
```

```python
sns.boxplot(x=df_iris["species"], y=df_iris["sepal_length"])
```

```python
sns.set_theme(style="whitegrid", palette="pastel")
sns.boxplot(x=df_iris["sepal_length"], y=df_iris["species"])
```

```python
# matplotlib ile boxplot
df_iris.boxplot(column='sepal_length', by='species')
plt.show()
```

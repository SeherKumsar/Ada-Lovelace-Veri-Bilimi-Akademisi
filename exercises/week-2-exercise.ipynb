**Understanding the data**
- 3 veriseti de 150 sütundan oluşmaktadır.
- Veriler 22 Ocak 2020'den itibaren başlayıp 15 Haziran 2020 tarihine kadar olan verileri içermektedir.
- Country/Region, Lat ve Long sütunları (latitude/longitude, enlem/boylam) verilerin hangi ülkeye ait olduğunu göstermektedir.
- Veri setlerindeki satırları incelediğimizde vaka ve ölüm verisetlerinin satır sayıları aynı iken, iyileşen verisetinde daha az satır olduğunu gözlemliyoruz.

```python
# Tekrar eden satırları kontrol et
print("Tekrar eden satırlar Enfekte olan veri setinde:")
total_duplicates_infected = infected_dataset.duplicated().sum()
print(total_duplicates_infected)

print("\nTekrar eden satırlar İyileşen veri setinde:")
total_duplicates_recovered = recovered_dataset.duplicated().sum()
print(total_duplicates_recovered)

print("\nTekrar eden satırlar Ölen veri setinde:")
total_duplicates_deaths = deaths_dataset.duplicated().sum()
print(total_duplicates_deaths)
```

```python
# Belirli bir sütundaki benzersiz değerlerin sayısını kontrol etmek için nunique() kullanılır
print("", infected_dataset['Country/Region'].nunique()) # Enfekte olan ülke sayısı
print("", recovered_dataset['Country/Region'].nunique()) # İyileşen ülke sayısı
print("", deaths_dataset['Country/Region'].nunique()) # Ölen ülke sayısı
```

**ANALYZE THE DATA SUCH AS `groupby()`**

```python
# Toplam vaka sayılarını ülkeler bazında göster
total_infected_by_country = infected_dataset.groupby("Country/Region").sum().iloc[:, 3:].max(axis=1)
print("\nToplam Vaka Sayıları Ülkeler Bazında:")
print(total_infected_by_country.sort_values(ascending=False))
```

### 

```python
# Toplam iyileşen sayılarını ülkeler bazında göster
total_recovered_by_country = recovered_dataset.groupby("Country/Region").sum().iloc[:, 3:].max(axis=1)
print("\nToplam İyileşen Sayıları Ülkeler Bazında:")
print(total_recovered_by_country.sort_values(ascending=False))
```

```python
# Toplam ölüm sayılarını ülkeler bazında göster
total_deaths_by_country = deaths_dataset.groupby("Country/Region").sum().iloc[:, 3:].max(axis=1)
print("\nToplam Ölüm Sayıları Ülkeler Bazında:")
print(total_deaths_by_country.sort_values(ascending=False))
```

**Total cases, recoveries and deaths by country**

```python
# Ülkeler bazında toplam vaka, iyileşen ve ölüm sayılarını göster
total_infected_recovered_deaths_by_country = pd.concat([total_infected_by_country, total_recovered_by_country, total_deaths_by_country], axis=1)
total_infected_recovered_deaths_by_country.columns = ["Toplam Vaka", "Toplam İyileşen", "Toplam Ölüm"]
print("\nÜlkeler Bazında Toplam Vaka, İyileşen ve Ölüm Sayıları:")
print(total_infected_recovered_deaths_by_country.sort_values(by="Toplam Vaka", ascending=False))
```

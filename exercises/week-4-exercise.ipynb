```python
def get_video_statistics(video_id, youtube_api_key):
    # collecting video statistics
    video_stats_url = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+youtube_api_key
    video_stats_response = requests.get(video_stats_url).json()

    view_count = video_stats_response['items'][0]["statistics"]["viewCount"]
    like_count = video_stats_response['items'][0]["statistics"]["likeCount"]

    return view_count, like_count
```

```python
def get_videos_from_channel(df, channel_id, youtube_api_key):
    # make API call
    pageToken = ""
    url = "https://www.googleapis.com/youtube/v3/search?key="+youtube_api_key+"&channelId="+channel_id+"&part=snippet,id&order=date&maxResults=1000"+pageToken
    
    response = requests.get(url).json()
    time.sleep(3)

    for video in response['items']:
        if video['id']['kind'] == 'youtube#video':
            video_title = video['snippet']['title']
            video_id = video['id']['videoId']
            upload_date = video['snippet']['publishedAt']
            upload_date= str(upload_date).split("T")[0]

            view_count, like_count = get_video_statistics(video_id, youtube_api_key)

            #save in pandas df
            df = df._append({
                "video_id": video_id,
                "video_title": video_title,
                "upload_date": upload_date,
                "view_count": view_count,
                "like_count": like_count
            }, 
            ignore_index = True
            )
    return df
```

```python
def get_stock_data(api_key, symbol):
    base_url = "https://www.alphavantage.co/query"
    function = "TIME_SERIES_DAILY"
    params = {"function": function, "symbol": symbol, "apikey": api_key}

    response = requests.get(base_url, params=params)
    data = response.json()

    if "Time Series (Daily)" in data:
        # Günlük kapanış fiyatlarını işleyin
        time_series = data["Time Series (Daily)"]
        for date, values in time_series.items():
            close_price = values["4. close"]
            print(f"{date}: Kapanış Fiyatı: {close_price}")
    else:
        print(f"Hisse Senedi Verileri API'sinden veri çekilemedi. Hata: {data.get('Note')}")
```

```python
def search_technology_stocks(api_key, keyword):
    search_url = f"https://www.alphavantage.co/query?function=SYMBOL_SEARCH&keywords={keyword}&apikey={api_key}"

    response = requests.get(search_url)
    search_data = response.json()

    if "bestMatches" in search_data:
        technology_stocks = search_data["bestMatches"]
        for stock in technology_stocks:
            symbol = stock['1. symbol']
            print(f"Symbol: {symbol}")
            get_stock_data(api_key, symbol)
            print("-" * 50)
    else:
        print("Teknoloji şirketleri bulunamadı.")
```

```python
def search_technology_stocks(api_key, keyword):
    search_url = f"https://www.alphavantage.co/query?function=SYMBOL_SEARCH&keywords={keyword}&apikey={api_key}"

    response = requests.get(search_url)
    search_data = response.json()

    if "bestMatches" in search_data:
        all_dataframes = []  # Her hisse senedi için bir DataFrame'i tutacak liste
        for stock in search_data["bestMatches"]:
            symbol = stock['1. symbol']
            print(f"Symbol: {symbol}")
            stock_dataframe = get_stock_data(api_key, symbol)
            all_dataframes.append(stock_dataframe)
            print("-" * 50)

        # Tüm DataFrame'leri birleştir
        if all_dataframes:
            result_df = pd.concat(all_dataframes, axis=1)
            print("\nToplam Veri Çerçevesi:")
            print(result_df)
        else:
            print("Hiçbir veri çerçevesi oluşturulamadı.")
    else:
        print("Teknoloji şirketleri bulunamadı.")

def get_stock_data(api_key, symbol):
    base_url = "https://www.alphavantage.co/query"
    function = "TIME_SERIES_DAILY"
    params = {"function": function, "symbol": symbol, "apikey": api_key}

    response = requests.get(base_url, params=params)
    data = response.json()

    if response.status_code != 200:
        print(f"Hata: {response.status_code}, {symbol} için veri çekilemedi.")
        return pd.DataFrame()

    if "Time Series (Daily)" not in data:
        print(f"{symbol} için günlük kapanış fiyatları bulunamadı. Hata: {data.get('Note')}")
        return pd.DataFrame()

    # DataFrame oluştur
    time_series = data["Time Series (Daily)"]
    df = pd.DataFrame(time_series).T  # Transpoze yaparak, tarihleri satırlara yerleştir
    df.index = pd.to_datetime(df.index)  # Tarih sütununu datetime formatına çevir
    df = df.rename(columns={"4. close": f"{symbol}_Close"})  # Sütun adını değiştir

    return df
```

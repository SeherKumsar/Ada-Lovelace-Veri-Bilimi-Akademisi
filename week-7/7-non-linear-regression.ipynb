## Polynomial Regression

y^= θ0 + θ1*x + θ2^*x^2 + θ3^*x^3

x1 = x
x2 = x^2
x3 = x^3

y^ = θ0 + θ1*x1 + θ2^*x2 + θ3^*x3

Polynominal regression is considered to be a special case of traditional multiple linear regression.

    - Polynominal regression models can fit using the model of "least squares"
    - Minimize the sum of the squares of the differences between y and y^ or (y_hat)

## Polynomial Regression
- non-linear relationship between the dependent variable x and a set of independent variables y
- For a model to be considered non-linear y^ must be a non-linear function of the parameters θ, not necessarily the features of x.
- in Non-linear regression, a model is non-linear by parameters.

How can I know if a problem is linear or non-linear?
- Visually figure out relation, use plots of output variables with each input variable.
- Calculate the correlation coefficient between independent and dependent variables.
    - if for all variable it is 0.7 or higher there is a linearity, and it's not appropriate to fit a non-linear regression.
